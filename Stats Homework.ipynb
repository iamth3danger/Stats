```python
#Bootstrap distribution
dist = []
for i in range(1000):
    dist.append(stats.mean(random.choices(samples, k = len(samples))))
```

```python
#list of means of distributions
dist
```

```python
#mean of entire distribution
few = dist
stats.mean(few)
```

```python
#standard error
scipy.stats.sem(few)
```

```python
rats_confidence95 = [stats.mean(few) - 1.96*scipy.stats.sem(few), stats.mean(few) + 1.96*scipy.stats.sem(few)]
rats_confidence90 = [stats.mean(few) - 1.645*scipy.stats.sem(few), stats.mean(few) + 1.645*scipy.stats.sem(few)]
rats_confidence99 = [stats.mean(few) - 2.576*scipy.stats.sem(few), stats.mean(few) + 2.576*scipy.stats.sem(few)]
```

```python
#95% confidence interval
rats_confidence95
```

```python
#90% confidence interval
rats_confidence90
```

```python
#99% confidence interval
rats_confidence99
```
